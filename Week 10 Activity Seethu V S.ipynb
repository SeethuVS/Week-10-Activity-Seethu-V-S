# a. ph and Chloramines

```python
data[['ph','Chloramines']].corr()
```

```python
data=data[['ph','Chloramines']]
corrmatrix=data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,annot=True)
```

```python
sns.scatterplot(data=data,x='Chloramines',y='ph')
plt.title('Chloramines vs ph')
plt.show()
```

We can notice that the correlation coefficient between the features Ph and chloramine = -0.024768 which is approximately equal to 0. Therefore, we can conclude there is zero correlation b/w ph and chloramine.
